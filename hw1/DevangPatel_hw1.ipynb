```python
import numpy as np 
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
```

```python
# read in the image
inputImg = plt.imread('image.jpg')
# set the image matrix to greyscale
inputImg = np.mean(inputImg, -1)
```

```python
plt.axis('off')
plt.imshow(inputImg)
```

```python
# SVD calculation
u, s, v = np.linalg.svd(inputImg)
# s is output as a 1D linear array so you have to make a diagonal array out of it
s = np.diag(s)
```

```python
for k in (1, 5, 10, 50, 100):
    newImg = u[:, :k] @ s[:k, :k] @ v[:k, :]
    plt.axis('off')
    plt.title('k = ' + str(k))
    plot = plt.imshow(newImg)
    plt.savefig('cat_k=' + str(k)+".jpg")
    plt.show()
    
```
